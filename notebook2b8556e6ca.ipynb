```python
class GbRegression:
    def __init__(self, l_rate, epochs):
        self.m = 95.22506146
        self.b = -120
        self.lr = l_rate
        self.e = epochs
        
    def fit(self, X, Y):
        for i in range(self.e):
            loss_slope = -2 * np.sum(Y - self.m * X.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            
        print(self.b)
```

```python
gdr=GbRegression(0.01,120)
```

```python
gdr.fit(X,y)
```
